```python
import os
import glob
import shutil
```

```python
root_dir = 'lia/datasets/vox/train'
videos = os.listdir(root_dir)
print(videos[0])
```

```python
cnt_total = 0
cnt_remove = 0
for i, video in enumerate(videos):
    video_path = os.path.join(root_dir, video)
    cnt = len(os.listdir(video_path))

    if cnt < 64:
        print(f'[{i}] cnt: {cnt}, {video_path}, removed')
        shutil.rmtree(video_path)
        cnt_remove += 1
    cnt_total += 1
print(f'{cnt_remove}/{cnt_total}')
```
